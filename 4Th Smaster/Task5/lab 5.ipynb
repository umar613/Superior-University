```python
gaussian_blur = cv2.GaussianBlur(resized_img, (15, 15), 0)
gaussian_rgb = cv2.cvtColor(gaussian_blur, cv2.COLOR_BGR2RGB)
plt.imshow(gaussian_rgb)
plt.title('Image with Gaussian Blur')
plt.axis('off')
plt.show()
```

2. Median Blur

```python
median_blur = cv2.medianBlur(resized_img, 11)
median_rgb = cv2.cvtColor(median_blur, cv2.COLOR_BGR2RGB)
plt.imshow(median_rgb)
plt.title('Image with Median Blur')
plt.axis('off')
plt.show()
```


3. Bilateral Blur

```python
bilateral_filtered = cv2.bilateralFilter(resized_img, 15, 150, 150)
bilateral_rgb = cv2.cvtColor(bilateral_filtered, cv2.COLOR_BGR2RGB)
plt.imshow(bilateral_rgb)
plt.title('Image with Bilateral Blur')
plt.axis('off')
plt.show()
```
